## 판다스(Pandas) 06 Page 3

```python
import pandas as pd

persons_dict = { 'Name' : [ 'John', 'Wei', 'Smith', 'Park', 'Lee' ],
                 'Country' : [ 'USA', 'China', 'USA', 'Korea', 'Korea' ],
                 'Age' : [ 31, 33, 28, 41, 55 ],
                 'Job' : ['Singer', 'Lawyer', 'Chef', 'Doctor', 'Chef'] } 

df1 = pd.DataFrame(data = persons_dict, 
                   columns = [ 'Name', 'Country', 'Age', 'Job' ] ) 

df1
```

```python
# 인덱스 0인 1개의 행 삭제
df1.drop( df1.index[0], inplace=True) 

df1
```

```python
# 불연속적인 행 삭제시 리스트 사용
df1.drop( df1.index[ [1, 3] ], inplace=True) 

df1
```

## 판다스(Pandas) 06 Page 4

```python
import pandas as pd

persons_dict = { 'Name' : [ 'John', 'Wei', 'Smith', 'Park', 'Lee' ],
                 'Country' : [ 'USA', 'China', 'USA', 'Korea', 'Korea' ],
                 'Age' : [ 31, 33, 28, 41, 55 ],
                 'Job' : ['Singer', 'Lawyer', 'Chef', 'Doctor', 'Chef'] } 

df1 = pd.DataFrame(data = persons_dict, 
                   columns = [ 'Name', 'Country', 'Age', 'Job' ] ) 

df1
```

```python
# 열 삭제, axis = 1 사용함
df1.drop( 'Job', axis=1, inplace=True ) 

df1
```

```python
# 여러개의 열을 삭재할 경우 리스트로 만들어서 사용함
df1.drop( ['Name', 'Age'], axis=1, inplace=True ) 

df1
```
