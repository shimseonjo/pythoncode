```python
import pandas  as pd
```

##### 1. 주민등록인구및세대현황.csv 가지고 와서 데이터 프레임으로 작성합니다.

```python
data = pd.read_csv("주민등록인구및세대현황.csv",encoding='cp949')
```

```python
data
```

##### 2. 인구정보 데이터 프레임의 컬럼명을 행정구역, 총인구수, 세대수, 세대당인구, 남자, 여자, 남녀비율 로 변경하세요

```python
data=data.rename(columns = {'2020년06월_총인구수':'총인구수', '2020년06월_세대수':'세대수', '2020년06월_세대당 인구':'세대당인구','2020년06월_남자 인구수': '남자', '2020년06월_여자 인구수':'여자', '2020년06월_남여 비율':'남여비율'})
          
```

```python
data.info()
```

```python
data.columns
```

##### 3. 남녀비율 컬럼을 제거합니다.

```python
data.drop('남여비율',axis=1, inplace=True)
```

```python
data.columns
```

##### 4. 행정구역 컬럼의 ‘(’ ‘)’를 제거하고 시,구,동 컬럼을 만들어 분리해 넣고 대구지역의 값만 가지고 와서 인덱스를 새로 부여합니다.

```python
data.행정구역=data.행정구역.str.split('(').str.get(0)
```

```python
data['시'] = data['행정구역'].str.split().str.get(0)
```

```python
data['구'] = data['행정구역'].str.split().str.get(1)
```

```python
data['동'] = data['행정구역'].str.split().str.get(2)
```

```python
data = data[data['시']=='대구광역시']
```

```python
data.reset_index()
```

##### 5. 총인구수, 남자, 여자 컬럼의 데이타 타입을 정수형으로 변경합니다.

```python
data['총인구수']=data['총인구수'].str.replace(',','').astype('int')
```

```python
data['남자']=data['남자'].str.replace(',','').astype('int')
```

```python
data['여자']=data['여자'].str.replace(',','').astype('int')
```

```python
data.info()
```
